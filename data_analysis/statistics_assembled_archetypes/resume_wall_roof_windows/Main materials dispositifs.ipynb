```python
def print_main_materials(dan_case, dpe_case):
    
    print("\nArchétypes Danube\n",dan_case.archetype.unique(),"\n")
    print("*"*60,"\n")

    mat_vit = ["adedpe202006_logtype_baie_type_vitrage",]
    mat_mur = ["cerffo2020_mat_mur_txt", "adedpe202006_logtype_mur_mat_ext", "adedpe202006_logtype_mur_ep_mat_ext",]
    mat_toit = ["adedpe202006_logtype_ph_mat"]

    # vitrage
    print("Vitrage Danube\n" ,
          dan_case.TYPE_VITRAGE.unique()[:3])
    print("_"*30,"\n")
    for mat_var in mat_vit:
    #     print(mat_var)
        print((dpe_case[mat_var].value_counts()[:3] / len(dpe_case) * 100).round(1),"\n")
    print("*"*60,"\n")

    # Murs
    print("Murs Danube - Options M1, M2, M2\n  PORTEUR" ,
          "\nM1", dan_case.EP_PORTEUR_M1.unique(), dan_case.PORTEUR_M1.unique(), 
          "\nM2", dan_case.EP_PORTEUR_M2.unique(), dan_case.PORTEUR_M2.unique(), 
          "\nM3", dan_case.EP_PORTEUR_M3.unique(), dan_case.PORTEUR_M3.unique(),)
    print("_"*30,"\n")
    for mat_var in mat_mur:
    #     print(mat_var)
        print((dpe_case[mat_var].value_counts()[:3] / len(dpe_case) * 100).round(1),"\n")
    print("*"*60,"\n")

    # Toit
    print("Toit Danube :" ,
          "\nT1", dan_case.EP_PORTEUR_T1.unique(), dan_case.PORTEUR_T1.unique(), 
          "\nT2", dan_case.EP_PORTEUR_T2.unique(), dan_case.PORTEUR_T2.unique(), )
    print("_"*30,"\n")
    for mat_var in mat_toit:
    #     print(mat_var)
        print((dpe_case[mat_var].value_counts()[:3] / len(dpe_case) * 100).round(1),"\n")
    print("#" * 100)
```

## generate table all archs available

```python
def table_main_materials(dan_case, dpe_case):
    print("Archétypes in csv file: ",dan_case.archetype.unique(),"\n")
    vars_mur = ["cerffo2020_mat_mur_txt", "adedpe202006_logtype_mur_mat_ext", "adedpe202006_logtype_mur_ep_mat_ext",]
    vars_toit = ["adedpe202006_logtype_ph_mat"]
    pop_taille = pd.Series(data = len(dpe_case), index=['Population (nb log.)'])

    # vitrage
    vit_dan = pd.Series(data = dan_case.TYPE_VITRAGE.unique()[0], index=['Vitrage Danube'])
    
    vit_dpe_data = dpe_case[dpe_case.adedpe202006_logtype_baie_type_vitrage.notnull()]["adedpe202006_logtype_baie_type_vitrage"]
    vit_samp = pd.Series(data = len(vit_dpe_data), index=['Echant. Vitrage DPE (nb log.)'])
    vit_dpe = {}
    mat_vit_dpe = (vit_dpe_data.value_counts()[:3] / len(vit_dpe_data) * 100).round(1)
    
    vit_dpe[f"{(mat_vit_dpe.index[0]).title()} DPE (%)"] = mat_vit_dpe.tolist()[0]
    vit_dpe[f"{(mat_vit_dpe.index[1].title())} DPE (%)"] = mat_vit_dpe.tolist()[1]
    if len(mat_vit_dpe) > 2:
        vit_dpe[f"{(mat_vit_dpe.index[2]).title()} DPE (%)"] = mat_vit_dpe.tolist()[2]
    vit_dpe_ser = pd.Series(data = vit_dpe)
#     print(vitrage)

    # Murs
    mur_dan = {}
    mur_dan["Mur Danube M1 (porteur)"] = dan_case.PORTEUR_M1.unique()[0]
    mur_dan["Mur Danube M2 (porteur)"] = dan_case.PORTEUR_M2.unique()[0]
    mur_dan["Mur Danube M3 (porteur)"] = dan_case.PORTEUR_M3.unique()[0]
    mur_dan_ser = pd.Series(data = mur_dan)
#     print(mur_dan_ser)

    mur_cerf_data = dpe_case[dpe_case.cerffo2020_mat_mur_txt.notnull()]["cerffo2020_mat_mur_txt"]
    mur_cerf_samp = pd.Series(data = len(mur_cerf_data), index=['Echant. Mur cerffo (nb log.)'])
    mur_cerf = {}
    mat_mur_cerf = (mur_cerf_data.value_counts()[:3] / len(mur_cerf_data) * 100).round(1)
  
    mur_cerf["Mur cerffo M1"] = mat_mur_cerf.index[0]
    mur_cerf["Mur cerffo M1 (%)"] = mat_mur_cerf.tolist()[0]

    mur_cerf["Mur cerffo M2"] = mat_mur_cerf.index[1]
    mur_cerf["Mur cerffo M2 (%)"] = mat_mur_cerf.tolist()[1]

    mur_cerf["Mur cerffo M3"] = mat_mur_cerf.index[2]
    mur_cerf["Mur cerffo M3 (%)"] = mat_mur_cerf.tolist()[2]
    mur_cerf_ser = pd.Series(data = mur_cerf)
#     print(mur_cerf_ser)

    mur_dpe_data = dpe_case[dpe_case.adedpe202006_logtype_mur_mat_ext.notnull()]["adedpe202006_logtype_mur_mat_ext"]
    mur_dpe_samp = pd.Series(data = len(mur_dpe_data), index=['Echant. Mur DPE (nb log.)'])
    mat_mur_dpe = (mur_dpe_data.value_counts()[:3] / len(mur_dpe_data) * 100).round(1)

    mur_dpe = {}
    mur_dpe["Mur DPE M1"] = mat_mur_dpe.index[0]
    mur_dpe["Mur DPE M1 (%)"] = mat_mur_dpe.tolist()[0]

    mur_dpe["Mur DPE M2"] = mat_mur_dpe.index[1]
    mur_dpe["Mur DPE M2 (%)"] = mat_mur_dpe.tolist()[1]

    mur_dpe["Mur DPE M3"] = mat_mur_dpe.index[2]
    mur_dpe["Mur DPE M3 (%)"] = mat_mur_dpe.tolist()[2]

    mur_dpe_ser = pd.Series(data = mur_dpe)
#     print(mur_dpe_ser)


    # # Toit
    toit_dan = {}
    toit_dan["Toit Danube T1 (porteur)"] = dan_case.PORTEUR_T1.unique()[0]
    toit_dan["Toit Danube T2 (porteur)"] = dan_case.PORTEUR_T2.unique()[0]
    toit_dan_ser = pd.Series(data = toit_dan)
#     print(toit_dan_ser)

    toit_dpe_data = dpe_case[dpe_case.adedpe202006_logtype_ph_mat.notnull()]["adedpe202006_logtype_ph_mat"]
    toit_dpe_samp = pd.Series(data = len(toit_dpe_data), index=['Echant. Toit DPE (nb log.)'])
    mat_toit_dpe = (toit_dpe_data.value_counts()[:3] / len(toit_dpe_data) * 100).round(1)
    
    toit_dpe = {}
    toit_dpe["Toit DPE T1"] = mat_toit_dpe.index[0]
    toit_dpe["Toit DPE T1 (%)"] = mat_toit_dpe.tolist()[0]

    toit_dpe["Toit DPE T2"] = mat_toit_dpe.index[1]
    toit_dpe["Toit DPE T2 (%)"] = mat_toit_dpe.tolist()[1]

    toit_dpe["Toit DPE T3"] = mat_toit_dpe.index[2]
    toit_dpe["Toit DPE T3 (%)"] = mat_toit_dpe.tolist()[2]

    toit_dpe_ser = pd.Series(data = toit_dpe)
#     print(toit_dpe_ser)

    mats = pd.concat([pop_taille,
                      vit_dan,
                      vit_samp,
                      vit_dpe_ser,
                      mur_dan_ser,
                      mur_cerf_samp,
                      mur_cerf_ser,
                      mur_dpe_samp,
                      mur_dpe_ser,
                      toit_dan_ser,
                      toit_dpe_samp,
                      toit_dpe_ser 
    ])

    print("\n",mats)
    return mats

# test function 
for csv in bounds.index[:2]:   
    path = source_folder / csv
    root_name = "-".join(path.stem.split("-")[2:-1]) 
    print("\n", root_name)
    print(csv)
    dpe = pd.read_csv(path)    
    dan_case = dan[dan.fichier == csv]
    table_main_materials(dan_case, dpe)
```

## Vitrage 

```python
def perc_vitrage(csv, dpe_case):
    vit_dict = {}
    dan_case = dan[dan.fichier == csv]
    print("\nArchétypes Danube\n",dan_case.archetype.unique(),"\n")

    counts = dpe_case["adedpe202006_logtype_baie_type_vitrage"].value_counts()
    simple = (counts["simple vitrage"] / len(dpe_case) * 100).round(1)
    double = (counts["double vitrage"] / len(dpe_case) * 100).round(1)
    if len(counts) > 2:
        triple = (counts["triple vitrage"] / len(dpe_case) * 100).round(1)
    else:
        triple = 0

    print(counts ,"\n")

    vit_dict["fichier"] = csv
    vit_dict["archétypes"] = list(dan_case.archetype.unique())

    vit_dict["simp_vit_3CL"] = simple
    vit_dict["doub_vit_3CL"] = double
    vit_dict["trip_vit_3CL"] = triple
    vit_dict["main_vit_3CL"] = counts.index[0]
    vit_dict["main_vit_Danube"] = dan_case.TYPE_VITRAGE.unique()[0].lower()

    return vit_dict
```

```python
perc_vitrage_files = []
for csv in bounds.index:   
    path = source_folder / csv
    root_name = "-".join(path.stem.split("-")[2:-1]) 
    print("\n", root_name)
    print(csv)
    
    dpe = pd.read_csv(path)    
    perc_vit_file = perc_vitrage(csv, dpe)
    
    perc_vitrage_files.append(perc_vit_file)
```
