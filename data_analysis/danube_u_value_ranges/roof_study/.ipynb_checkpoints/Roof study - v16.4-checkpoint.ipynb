```python
# exclude cases with territoire France global
archs_nofr = [arch for arch in archs 
              if arch.split("-")[-1] != "FRANCE"]
len(archs_nofr)
```

```python
# Take out the ceiling from the archetypes

start = ["-".join(arch.split("-")[:-1]) 
         for arch in archs_nofr
         ]
print(start[0])

ter_mur = ["_".join(arch.split("-")[-1].split("_")[:-1])
            for arch in archs_nofr 
          ]
print(ter_mur[0])

n = range(len(start))
arch_ter_murs = ['-'.join([start[i], ter_mur[i]]) for i in n]
arch_ter_murs[0]
```

```python
# Take the set of the list

print(len(arch_ter_murs))
arch_ter_murs = set(arch_ter_murs)
print(len(arch_ter_murs))
```

```python
# Group archetypes in pairs or group of three

two_ceilings = []
three_ceilings =[]
for arch_mur in arch_ter_murs:
    same_mur = []
    for arch in archs_nofr:
#         if arch.split("-")[-1] != "FRANCE":
        if arch_mur in arch:
            same_mur.append(arch)
    if len(same_mur) == 2:
        two_ceilings.append(same_mur)
    elif len(same_mur) > 2:
        three_ceilings.append(same_mur)
        
print("len two_ceilings is ", len(two_ceilings)) 
print("len three_ceilings is ", len(three_ceilings)) 

print("\nthree_ceilings:\n",three_ceilings)
print("\ntwo_ceilings\n",two_ceilings)
```

```python
# print cases in which the roofs have different properties (besides material)
for pair in two_ceilings:
    print('\n', "-"*100)
    for e in pair:
        print(e)
    
    df0 = df[df["arch"]==pair[0]]
    df1 = df[df["arch"]==pair[1]]
    pair_df = pd.concat([df0,df1])[["TERRITOIRE",
                                    "REVETEMENT_EXT_T1",
                                    "EP_RE_T1",
                                    "PORTEUR_T1",
                                    "EP_PORTEUR_T1",
                                    "ISOLANT_T1",
                                    "EP_ISOLANT_T1",
                                    "REVETEMENT_INT_T1",
                                    "EP_RI_T1",
                                    ]]
    if sum(pair_df.iloc[1][2:] != pair_df.iloc[0][2:]) != 0:
        display(pair_df)
```

```python
for group in three_ceilings:
    print('\n', "-"*100)
    for e in group:
        print(e)
    
    df0 = df[df["arch"]==group[0]]
    df1 = df[df["arch"]==group[1]]
    df2 = df[df["arch"]==group[2]]

    group_df = pd.concat([df0,df1,df2])[["TERRITOIRE",
                                    "REVETEMENT_EXT_T1",
                                    "EP_RE_T1",
                                    "PORTEUR_T1",
                                    "EP_PORTEUR_T1",
                                    "ISOLANT_T1",
                                    "EP_ISOLANT_T1",
                                    "REVETEMENT_INT_T1",
                                    "EP_RI_T1",
                                    ]]
    if (sum(group_df.iloc[0][2:] != group_df.iloc[1][2:]) + 
        sum(group_df.iloc[0][2:] != group_df.iloc[2][2:])
        +
        sum(group_df.iloc[1][2:] != group_df.iloc[2][2:])
           )!= 0:
        display(group_df)
```

```python
### Test in one column: applying the values of the table Uph_inconnu, to a new column
# disp_roof['Uph_inc_H1_joule'] = disp_roof['period'].map(dict_isol_inc["H1_comble_joule"])
```

```python
                             
dict_comble = dict(zip(name_Uph_inc,inc_comble))
dict_terrace = dict(zip(name_Uph_inc,inc_terrace))

for name_col in name_Uph_inc:
    # calculate based on the table and on the type of roof (combles/terraces)                    
    disp_roof[name_col] = np.where(cond_bois, 
                                    disp_roof['period'].map(dict_isol_inc[dict_comble[name_col]]),
                                    disp_roof['period'].map(dict_isol_inc[dict_terrace[name_col]])
                                    )

    # correct P6 in the "autre" using P7 values
    if name_col.endswith("autre"):
        cond_P6 = disp_roof["period"] == 'P6'

        disp_roof[name_col] = np.where((cond_P6 & cond_bois), # for combles
                                    dict_isol_inc[dict_comble[name_col]]['P7'], # if: change to P7 value
                                    disp_roof[name_col]) # else: just let the value which was before

        disp_roof[name_col] = np.where((cond_P6 & ~cond_bois), # for terraces
                                dict_isol_inc[dict_terrace[name_col]]['P7'],
                                disp_roof[name_col])

    # erase the values for serre
    disp_roof[name_col] = np.where(cond_serre, 
                                   np.nan,
                                   disp_roof[name_col])                                  
```

#### Uph_R

```python
# generalization
def calculate_Uph_R(row, name_col, dict_cols):
    if row['EP_ISOLANT_T1'] > 0:
        if row['ISOLANT_T1'] == "ITI_LM":
            conduct = 0.04
        if row['ISOLANT_T1'] == 'ITE_POL':
            conduct = 0.03
        Uph_R = 1 / ( (1 / row[dict_cols[name_col]] ) + (row['EP_ISOLANT_T1'] / 100) / conduct)

        return Uph_R
    else:
        return np.nan
```

```python
name_Uph_R = ["Uph_R_ah",
            "Uph_R_b",
            "Uph_R_c",
            "Uph_R_de",
            "Uph_R_fgij"]

dict_cols_R = dict(zip(name_Uph_R,name_Uph_0))

for name_col in name_Uph_R:
    disp_roof[name_col] = disp_roof.apply(lambda x : calculate_Uph_R(x, name_col, dict_cols_R), axis=1)
```

#### Uph_E

```python
def calculate_Uph_E(row, name_col, dict_cols):
    if row['EP_ISOLANT_T1'] > 0:
        Uph_E = 1 / ( (1 / row[dict_cols[name_col]] ) + (row['EP_ISOLANT_T1'] / 100) / 0.04)

        return Uph_E
    else:
        return np.nan
```

```python
name_Uph_E = ["Uph_E_ah",
            "Uph_E_b",
            "Uph_E_c",
            "Uph_E_de",
            "Uph_E_fgij"]

dict_cols_E = dict(zip(name_Uph_E,name_Uph_0))

for name_col in name_Uph_E:
    disp_roof[name_col] = disp_roof.apply(lambda x : calculate_Uph_E(x, name_col, dict_cols_E), axis=1)
```
