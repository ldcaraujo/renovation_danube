```python
df["arch"] = (df['PERIODE'] + "-" + 
              df['NOM_TYPOLOGIE'] + "-" +
              df['USAGE'] + "-" +
              df['TERRITOIRE'])
archs = df["arch"].unique()
```

```python
# exclude cases with territoire France global
archs_nofr = [arch for arch in archs 
              if arch.split("-")[-1] != "FRANCE"]
len(archs_nofr)
```

```python
# Take out the ceiling from the archetypes

start = ["-".join(arch.split("-")[:-1]) 
         for arch in archs_nofr
         ]
print(start[0])

ter_mur = ["_".join(arch.split("-")[-1].split("_")[:-1])
            for arch in archs_nofr 
          ]
print(ter_mur[0])

n = range(len(start))
arch_ter_murs = ['-'.join([start[i], ter_mur[i]]) for i in n]
arch_ter_murs[0]
```

```python
# Take the set of the list

print(len(arch_ter_murs))
arch_ter_murs = set(arch_ter_murs)
print(len(arch_ter_murs))
```

```python
# Group archetypes in pairs or group of three

two_ceilings = []
three_ceilings =[]
for arch_mur in arch_ter_murs:
    same_mur = []
    for arch in archs_nofr:
#         if arch.split("-")[-1] != "FRANCE":
        if arch_mur in arch:
            same_mur.append(arch)
    if len(same_mur) == 2:
        two_ceilings.append(same_mur)
    elif len(same_mur) > 2:
        three_ceilings.append(same_mur)
        
print("len two_ceilings is ", len(two_ceilings)) 
print("len three_ceilings is ", len(three_ceilings)) 

print("\nthree_ceilings:\n",three_ceilings)
print("\ntwo_ceilings\n",two_ceilings)
```

```python
# print cases in which the roofs have different properties (besides material)
for pair in two_ceilings:
    print('\n', "-"*100)
    for e in pair:
        print(e)
    
    df0 = df[df["arch"]==pair[0]]
    df1 = df[df["arch"]==pair[1]]
    pair_df = pd.concat([df0,df1])[["TERRITOIRE",
                                    "REVETEMENT_EXT3",
                                    "EP_RE3",
                                    "PORTEUR3",
                                    "EP_PORTEUR3",
                                    "ISOLANT3",
                                    "EP_ISOLANT3",
                                    "REVETEMENT_INT3",
                                    "EP_RI3",
                                    ]]
    if sum(pair_df.iloc[1][2:] != pair_df.iloc[0][2:]) != 0:
        display(pair_df)
```

```python
for group in three_ceilings:
    print('\n', "-"*100)
    for e in group:
        print(e)
    
    df0 = df[df["arch"]==group[0]]
    df1 = df[df["arch"]==group[1]]
    df2 = df[df["arch"]==group[2]]

    group_df = pd.concat([df0,df1,df2])[["TERRITOIRE",
                                    "REVETEMENT_EXT3",
                                    "EP_RE3",
                                    "PORTEUR3",
                                    "EP_PORTEUR3",
                                    "ISOLANT3",
                                    "EP_ISOLANT3",
                                    "REVETEMENT_INT3",
                                    "EP_RI3",
                                    ]]
    if (sum(group_df.iloc[0][2:] != group_df.iloc[1][2:]) + 
        sum(group_df.iloc[0][2:] != group_df.iloc[2][2:])
        +
        sum(group_df.iloc[1][2:] != group_df.iloc[2][2:])
           )!= 0:
        display(group_df)
```
