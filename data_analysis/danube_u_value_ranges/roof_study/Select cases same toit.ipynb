```python
import os
```

```python
folder = r"C:\Users\ldecarva\Documents\Data\gpd_sorting_output\archetypes_concatenated"
```

```python
files = os.listdir(folder)
files
```

```python
# Take out the ceiling from the archetypes

start = ["-".join(file.split("-")[1:-1]) for file in files]
print(start[0])

ter_mur = ["_".join(file.split("-")[-1].split("_")[:-1]) for file in files]
print(ter_mur[0])

n = range(len(start))
arch_ter_murs = ['-'.join([start[i], ter_mur[i]]) for i in n]
arch_ter_murs[0]
```

```python
# Take the set of the list

print(len(arch_ter_murs))
arch_ter_murs = set(arch_ter_murs)
print(len(arch_ter_murs))
```

```python
# Group archetypes in pairs or group of three

two_ceilings = []
three_ceilings =[]
for arch_mur in arch_ter_murs:
    same_mur = []
    for file in files:
        if arch_mur in file:
            same_mur.append(file)
    if len(same_mur) == 2:
        two_ceilings.append(same_mur)
    elif len(same_mur) > 2:
        three_ceilings.append(same_mur)
        
print("len two_ceilings is ", len(two_ceilings)) 
print("len three_ceilings is ", len(three_ceilings)) 

print("\nthree_ceilings:\n",three_ceilings)
print("\ntwo_ceilings\n",two_ceilings)
```

```python
# select archetypes to analyse according to the sample size differences

to_analyse = [] 
for pair in two_ceilings:   
    a = int(pair[0].split("-")[0].split("_")[1])
    b = int(pair[1].split("-")[0].split("_")[1])
    dif = abs(b-a)
    if (min(a,b) > 500 ) and(2 * dif < min(a,b)):
        print("\nBoth passed criteria")
        print("dif: ", dif, " minimun: ", min(a,b), "\n", pair)
        to_analyse.append(pair)


for group in three_ceilings:   
    a = int(group[0].split("-")[0].split("_")[1])
    b = int(group[1].split("-")[0].split("_")[1])
    c = int(group[2].split("-")[0].split("_")[1])
    dif_ab = abs(a-b)
    dif_bc = abs(b-c)
    dif_ca = abs(c-a)
    dif = max(dif_ab, dif_bc, dif_ca)
    mini = min(a, b, c)
    if (mini > 2000 ) and( dif < 100 * mini):
        print("\n3 passed criteria")
        print("dif: ", dif, " minimun: ", mini, "\n", group)
        to_analyse.append(group)    
    
```
