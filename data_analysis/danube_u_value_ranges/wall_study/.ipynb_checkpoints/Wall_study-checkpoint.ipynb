The combinations of "PORTEUR" and "REVETEMENT_EXT", presented in the previous part, were associated to the 3CL construction techniques.

#### Uw_R

```python
# generalization
def calculate_Uw_R(row, name_col, dict_cols):
    if row['EP_ISOLANT'] > 0:
        if (row['ISOLANT'] == "ITI_LM") or (row['ISOLANT'] == "ITE_LM"):
            conduct = 0.04
        if row['ISOLANT'] == 'ITE_POL':
            conduct = 0.03
        if row['ISOLANT'] == 'ITI_AIR':
            conduct = 0.025
        Uw_R = 1 / ( (1 / row[dict_cols[name_col]] ) + (row['EP_ISOLANT'] / 100) / conduct)

        return Uw_R
    else:
        return np.nan
```

```python
dict_cols_R = {"Uw_R":"U_0"}

disp_wall["Uw_R"] = disp_wall.apply(lambda x : calculate_Uw_R(x, "Uw_R", dict_cols_R), axis=1)
disp_wall
```

```python
disp_wall[disp_wall["ISOLANT"]=='ITI_AIR']
```

#### Uw_E

```python
def calculate_Uw_E(row, name_col, dict_cols):
    if row['EP_ISOLANT'] > 0:
        Uw_E = 1 / ( (1 / row[dict_cols[name_col]] ) + (row['EP_ISOLANT'] / 100) / 0.04)

        return Uw_E
    else:
        return np.nan
```

```python
dict_cols_E = {"Uw_E":"U_0"}

disp_wall["Uw_E"] = disp_wall.apply(lambda x : calculate_Uw_E(x, "Uw_E", dict_cols_E), axis=1)
```
