```python
def ceil_U_max(num):
    if math.isnan(num):
        return num
    else:
        return math.ceil(num * 100) / 100

num = 0.1714285714285714
# num= None
print(ceil_U_max(num))
```

```python
def floor_U_min(num):
    if math.isnan(num):
        return num
    else:
        return math.floor(num * 100) / 100

num = 0.1714285714285714
# num= np.nan
print(floor_U_min(num))
```

### U boundaries files  

```python
pd.set_option('display.max_colwidth', None)
```

```python
df_anlz = df[df['fichier'].notnull()]
```

```python
file_gb = df_anlz[['archetype', 'fichier', 
                     'Uph_min', 
                     'Uph_max', 
                     'Uw_min', 
                     'Uw_max',
                    ]].groupby("fichier")
```

```python
min_max_file = file_gb.agg({'Uw_min' : [np.min, np.max],
            'Uw_max' : [np.min, np.max],
            'Uph_min' : [np.min, np.max],
            'Uph_max' : [np.min, np.max]})
```

```python
min_max_file[("Uw_min_diff")] = min_max_file[("Uw_min","amin")] - min_max_file[("Uw_min","amax")]
min_max_file[("Uw_max_diff")] = min_max_file[("Uw_max","amin")] - min_max_file[("Uw_max","amax")]

min_max_file[("Uph_min_diff")] = min_max_file[("Uph_min","amin")] - min_max_file[("Uph_min","amax")]
min_max_file[("Uph_max_diff")] = min_max_file[("Uph_max","amin")] - min_max_file[("Uph_max","amax")]

min_max_file
```
