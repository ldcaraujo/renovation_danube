```python
from pathlib import Path
from os import listdir
import pandas as pd
```

```python
# charge archetypes from dept archs outputs
# arch_per_dept_paths = Path(__file__).parent.parent.parent.parent / "dados" / "_output_code" /"archetypes_per_dept_tables"
arch_per_dept_paths = r"C:\Users\laraujo\Music\Work_august\dados\_output_code\archetypes_per_dept_tables"
files_archs_depts = listdir(arch_per_dept_paths)

print(files_archs_depts[:2])
archs_depts = ["-".join(file_name.split("-")[:4]) for file_name in files_archs_depts]
print(f"total of archs-depts {len(archs_depts)}")
archs_depts = set(archs_depts)
print(f"unique archs_depts len of {len(set(archs_depts))}")
archs_depts
```

# Comparison Danube and archs_depts

```python
# charge Danube data base
danube_path = r"C:\Users\laraujo\Music\Work_august\dados\Danube\20220616_BddMAPUCE_MAP_V16.4-draft-export-utf8.csv"
danube = pd.read_csv(danube_path, sep=";")
# create column with Danube archetypes
danube["archetype"] = (danube['PERIODE'].astype(str) + "-" +
                    danube['NOM_TYPOLOGIE'].astype(str) + "-" +
                    danube['USAGE'].astype(str) + "-" +
                    danube['TERRITOIRE'].astype(str))
# list of all Danube archetypes
danube_archs = list(danube["archetype"])
print(danube_archs[:2])

print(len(danube_archs))
print(f"danube_archs len of {len(danube_archs)}")
print(len(set(danube_archs)))
```

## Territory

### P1

```python
# territories from danube for period P1
ter_p1 = danube[danube['PERIODE']=="P1"]["TERRITOIRE"].unique()
print(len(ter_p1))
list(ter_p1)
```

```python
# Territory France is not included !
archs_depts_p1 = [arch for arch in archs_depts if "P1" in arch]
archs_depts_p1
terr_archs_depts_p1 = [arch_p1.split("-")[-1] for arch_p1 in archs_depts_p1]
terr_archs_depts_p1 = set(terr_archs_depts_p1)
print(len(terr_archs_depts_p1))
terr_archs_depts_p1
```

## Intersection

```python
intersection = set(danube_archs) & set(archs_depts)
print(f"intersection len of {len(intersection)}")
intersection
```

```python
# archetypes in danube but not in arhcs_depts
archs_just_danube = set(danube_archs) - intersection
print(f"archs_just_danube len of {len(archs_just_danube)}")
archs_just_danube
```

```python
# OK, lists are complementary
len(archs_just_danube) + len(intersection)  - len(danube_archs)
```

```python
# archetypes in arhcs_depts but not in danube
archs_just_depts = set(archs_depts) - intersection
print(f"archs_just_depts len of {len(archs_just_depts)}")
archs_just_depts
```

```python
# OK, lists are complementary
len(archs_just_depts) + len(intersection) - len(archs_depts)
```
