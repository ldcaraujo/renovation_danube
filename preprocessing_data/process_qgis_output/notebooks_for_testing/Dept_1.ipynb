### Usage

```python
relations = {           
             'BATIMENT INDUSTRIEL' :
                 (dep1['cerffo2020_usage_niveau_2_txt'] == 'Industrie' )
                 | (dep1['cerffo2020_usage_niveau_3_txt'] == 'Etablissement industriel autre que carrière' ),       
             
             'SERRE AGRICOLE':
                 (dep1['cerffo2020_usage_niveau_3_txt'] == 'Serre' ),
             
             'BATIMENT D ENSEIGNEMENT':
                 (dep1['cerffo2020_usage_niveau_2_txt'] == 'Enseignement' ),
             
             'BATIMENT DE SANTE':
                 (dep1['cerffo2020_usage_niveau_2_txt'] == 'Centre de santé' ),
                         
             'COMMERCE':
                 (dep1['cerffo2020_usage_niveau_2_txt'] == 'Commerce' ),
             
             'HABITAT':
                 (dep1['cerffo2020_usage_niveau_1_txt'] == 'Résidentiel individuel' )
                 | (dep1['cerffo2020_usage_niveau_1_txt'] == 'Résidentiel collectif' ) # not described in BDNB methodology
                 | (dep1['cerffo2020_usage_niveau_3_txt'] == 'Maison exceptionnelle' ),
             
             'TERTIAIRE':
                 (dep1['cerffo2020_usage_niveau_2_txt'] == 'Bureau' )
                 | (dep1['cerffo2020_usage_niveau_2_txt'] == 'Hôtel' ),      
                          
             'BATIMENT SPORTIF':
                 (dep1['cerffo2020_usage_niveau_3_txt'] == 'Espace sportif' )
                 | (dep1['cerffo2020_usage_niveau_3_txt'] == 'Espace loisir' ),}

Usages_ignored = [
    'BATIMENT AGRICOLE',
    'BATIMENT RELIGIEUX',
    'CHATEAU',
    'LOCAL NON CHAUFFE'
]
```

```python
# create column with usages according to the criteria above
usages = list(relations.keys())
conditions = list(relations.values())

dep1['USAGE'] = np.select(conditions, usages, default=pd.NA)
```
